```python
#Importamos librerías requeridas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data frame del 2022 (Se eliminaron las filas que contenían solo valores nulos e impedían la salida de los índices)

df_vistazo= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2022', skiprows = 3)
df_vistazo
```

```python
#Verificamos información del DataFrame
df_vistazo.info()
```

```python
# Se identifican los valores nulos
valores_nulos=df_vistazo.isnull().sum()
valores_nulos
```

```python
#Analizar categorias de columnas con ceros

df_vistazo.groupby(['RET IVA'])['RET IVA'].count()
df_vistazo.groupby(['RET ISR'])['RET ISR'].count()
```

```python
# Eliminamos la columna de Poliza ya que la mayoría de datos son faltantes y no hay un patrón identificable para rellenar
# Se eliminan las variables de ISR y de IVA ya que la mayoría son ceros representando más del 90% de los datos.

df_vistazo2= df_vistazo.drop(['RET IVA','RET ISR','Otros'], axis=1)
df_vistazo2
```

```python
# Se sustituyó el folio por un registro único de identificación para tener un Data Frame Limpio al comenzar la observación de Outliers

df_vistazo2["Folio"] = df_vistazo2["Folio"].fillna("10000000000")
df_vistazo2["MP"] = df_vistazo2["MP"].fillna("Desconocido")
df_vistazo2["Poliza"] = df_vistazo2["Poliza"].fillna("Desconocido")

# Se sustituyó la variable "TC" por la mediana ya que la mayoría de las variables tienen el mismo número (1)

df_vistazo2["TC"]= df_vistazo2["TC"].fillna(round(df_vistazo2["TC"].median(),1))

df_vistazo2
```

```python
valores_nulos2=df_vistazo2.isnull().sum()
valores_nulos2
```

```python
#Filtro para contar solo con variables cualitativas y excluirlas de los métodos outliers
df_cuali = df_vistazo2.iloc[ : , [0,1,2,3,4,5,6,7,13,14,15]]
df_cuali
```

```python
# Filtro para solo contar con variables cuantitativas
df_cuanti= df_vistazo2.drop(['Fecha','Folio','UUID','RFC','Proveedor','TIPO GASTO','Descripción','MP','Tipo','Status','Poliza'], axis=1)
df_cuanti
```

```python
# Se identifican los valores nulos de las variables cuantitativas
valores_nulos3=df_cuanti.isnull().sum()
valores_nulos3

# Al tener nulos ya se pueden trabajar los outliers
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df_cuanti.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() # Se dibuja el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y = df_cuanti

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe

outliers_iqr= df_cuanti[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= df_cuanti[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
# Se corroboran los valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data4_iqr = data3_iqr.copy()
data4_iqr
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "median" para la variable de "TC" ya que la mayoría de datos tienen un mismo valor

data4_iqr["TC"]= data4_iqr["TC"].fillna(round(data3_iqr["TC"].median(),1))

# Reemplazamos valores atípicos (nulos) del dataframe con "mean" para las variables enteramente monetarias para no desvirtuar el resultado de los outliers

data4_iqr["Importe"]= data4_iqr["Importe"].fillna(round(data3_iqr["Importe"].mean(),1))
data4_iqr["IVA"]= data4_iqr["IVA"].fillna(round(data3_iqr["IVA"].mean(),1))
data4_iqr["TOTAL MX"]= data4_iqr["TOTAL MX"].fillna(round(data3_iqr["TOTAL MX"].mean(),1))
data4_iqr["TOTAL SAT"]= data4_iqr["TOTAL SAT"].fillna(round(data3_iqr["TOTAL SAT"].mean(),1))
```

```python
# Se identifican los valores nulos de las variables cuantitativas
valores_nulos4=data4_iqr.isnull().sum()
valores_nulos4
```

```python
# Unimos las variables cualitativas a las cuantitativas con el dataframe

Datos_limpios = pd.concat([df_cuali, data4_iqr], axis=1)
Datos_limpios
```

```python
# Verificación de los valores nulos del dataframe LIMPIO
valores_nulos_Final = Datos_limpios.isnull().sum()
valores_nulos_Final
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios_2022.csv")
```
